# Łączenie z bazą SQL



```python
import pandas as pd
```

uniwersalny sposób dla różnych baz danych

Musimy się połączyć z naszą bazą danych. Do tego celu musimy znać typ bazy, nazwę użytkownika i hasło, localhost, port i nazwę bazy. W dokumentacji widzimy jak połączyć te informacje w zależności od naszej bazy:      
https://docs.sqlalchemy.org/en/latest/core/engines.html

Aby uzyskać dostęp do bazy i powyższe informacje skontaktuj się odpowiednim działem w twojej organizacji.

```python
from sqlalchemy import create_engine
# see sqlalchemy docs for how to write this url for your database type:
```

```python
engine = create_engine('postgresql+psycopg2://postgres:haslo@localhost:5432/dvdrental')
```

```python
engine.table_names()
```

```python
pd.read_sql("select * from film where release_year = 2006", engine)
```

```python
table_name = 'actor'
df = pd.read_sql(table_name, engine)
```

```python
# Wysyłamy query do bazy i jako wynik otrzymujemy obiekt DataFrames
df = pd.read_sql("SELECT last_name, first_name FROM actor", engine)
```

```python
df.to_sql("nowa_tabela2", engine)
```

```python
# Tabele w naszej bazie
engine.table_names()
```

```python
# Zapisujemy do bazy SQL (pod warunkiem że mamy uprawnienia do tworzenia nowych tabel)
df.to_sql("nowa nowa nowa",engine)
```

```python
engine.table_names()
```

```python
# Wczytujemy całą tabelę
pd.read_sql("nowa nowa nowa", engine)
```

### Wykonujemy query bez fetch (bez pobierania danych do pythona)

```python
engine.execute("SELECT * FROM actor")
```

```python
engine.execute("DROP TABLE nowa_tabela2")
```
